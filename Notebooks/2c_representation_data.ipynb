## Archivo de datos: **Vehiculos_afectos_a_IPV_Ene-Mar_2025_3.xlsx**


```python
df = pd.read_excel(f"{working_dir}/Vehiculos_afectos_a_IPV_Ene-Mar_2025_3.xlsx")
print(df.shape)
df.head()
```

```python
df.rename(columns={
    'N° registro': 'num_registro',
    'Nº persona': 'num_persona',
    'Tipo persona': 'tipo_persona',
    'Tipo contribuyente': 'tipo_contribuyente',
    'Distrito domicilio fiscal': 'distrito_fiscal',
    'Porcentaje propiedad': 'porcentaje_propiedad',
    'Correlativo': 'correlativo',
    'Fecha adquisición vehículo': 'fecha_adquisicion',
    'Fecha 1ra inscripción SUNARP': 'fecha_inscripcion_sunarp',
    'Clase vehículo': 'clase_vehiculo',
    'Categoría vehículo': 'categoria_vehiculo',
    'Marca': 'marca',
    'Modelo': 'modelo',
    'Valor adquisición (US $)': 'valor_adquisicion_usd',
    'Valor adquisición (S/)': 'valor_adquisicion_pen',
    'Valor referencial 2025 (S/)': 'valor_referencial_2025_pen'
}, inplace=True)
df.head()
```

```python
# Omitimos duplicados
print(df.shape)
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df.dtypes
```

## Representación de Variables **Categóricas (con pocas categorías)**

```python
tab1 = df[['tipo_persona','num_registro']].groupby(['tipo_persona']).count().reset_index()
tab1
```

```python
tab2 = df[['tipo_persona', 'clase_vehiculo', 'num_registro']].groupby(['tipo_persona', 'clase_vehiculo']).count().reset_index()
tab2
```

## Representación de Variables **Categóricas (con al menos 1v de varias categorías)**

```python
tab1 = df[['marca','num_registro']].groupby(['marca']).count().reset_index().sort_values('num_registro', ascending=False)
tab1.head(10)
```

```python
tab2 = df[['marca', 'clase_vehiculo','num_registro']].groupby(['marca', 'clase_vehiculo']).count().reset_index()
tab2_top = tab2[tab2['marca'].isin(tab1.head(10)['marca'].values)]
tab2_top
```

```python
counts = tab2_top.groupby(['marca', 'clase_vehiculo']).sum().unstack(fill_value=0)
counts.columns = counts.columns.get_level_values(-1)
counts
```

```python
import pandas as pd
import matplotlib.pyplot as plt

ax = counts.plot(kind='bar', width=0.8, figsize=(12, 6))
plt.title('Distribución de las Top 10 Marcas por Clase de Vehículo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend(title='Clase vehículo')

for container in ax.containers:
    ax.bar_label(container, padding=1, fontsize=8)

plt.tight_layout()
plt.show()
```

## Representación de Variables **Numéricas Continuas**

```python
import seaborn as sns

#Histogramas por Tipo persona
sns.histplot(data=df[df['tipo_persona'] == 'Persona Natural'], x='valor_adquisicion_usd', color='orange', label='Natural', kde=False, bins=30)
sns.histplot(data=df[df['tipo_persona'] == 'Persona Jurídica'], x='valor_adquisicion_usd', color='green', label='Jurídica', kde=False, bins=30)
plt.legend()
plt.title('Distribución del Valor de Adquisición (US$) por Tipo de Persona')

# Custom formatter for abbreviated labels
def abbreviate(x, pos):
    if x >= 1e6:
        return f'{x/1e6:.1f}M'
    elif x >= 1e3:
        return f'{x/1e3:.0f}K'
    else:
        return f'{x:.0f}'

plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(abbreviate))

plt.show()
```

```python
import seaborn as sns

# Ojiva para datos por Tipo persona
sns.histplot(data=df[df['tipo_persona'] == 'Persona Natural'], x='valor_adquisicion_usd', color='orange', label='Natural', cumulative=True, bins=30, stat='density')
sns.histplot(data=df[df['tipo_persona'] == 'Persona Jurídica'], x='valor_adquisicion_usd', color='green', label='Jurídica', cumulative=True, bins=30, stat='density')

plt.legend()
plt.title('Distribución Acumulativa del Valor de Adquisición (US$)')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(abbreviate))
plt.show()
```

```python
import seaborn as sns

#Densidad
sns.kdeplot(data=df.loc[df['tipo_persona'] == 'Persona Natural'], x='valor_adquisicion_usd', color="orange", label="Natural")
sns.kdeplot(data=df.loc[df['tipo_persona'] == 'Persona Jurídica'], x='valor_adquisicion_usd', color="green", label="Jurídica")
plt.legend()
plt.title("Distribución de Valor de Adquisición por Tipo de Persona")
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(abbreviate))
plt.show()
```

```python
print( "Media de Valor Adquisición (US$) - Natural:", df.loc[df['tipo_persona'] == 'Persona Natural'][['valor_adquisicion_usd']].mean().values )
print( "Media de Valor Adquisición (US$) - Jurídica:", df.loc[df['tipo_persona'] == 'Persona Jurídica'][['valor_adquisicion_usd']].mean().values )
```

```python
print( "Mediana de Valor Adquisición (US$) - Natural:", df.loc[df['tipo_persona'] == 'Persona Natural'][['valor_adquisicion_usd']].median().values )
print( "Mediana de Valor Adquisición (US$) - Jurídica:", df.loc[df['tipo_persona'] == 'Persona Jurídica'][['valor_adquisicion_usd']].median().values )
```
