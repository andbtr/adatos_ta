### Análisis variable "modelo"

La columna modelo tiene 3,775 categorías únicas, la gran mayoría con muy pocos registros. Mantenerla como variable categórica generaría ruido, aumentaría el tiempo de cómputo y podría distorsionar los clusters. Por ello, se elimina, conservando otras variables que capturan información relevante del vehículo (marca, clase, categoría) para el análisis de clustering.

Se podría reducir a las 20–30 categorías más frecuentes, pero incluso así los modelos raros seguirían siendo numerosos y dispersos, lo que complicaría la interpretación y no aportaría información significativa adicional. Por eso se decidió eliminar la variable y mantener otras columnas relevantes para caracterizar los vehículos.

```python
df["modelo"].value_counts().head(10)
```

### Análisis variable "marca"

La variable marca tiene 133 categorías, pero muchas son muy poco frecuentes, incluso con solo 1 registro. Para mejorar la estabilidad y la interpretabilidad del clustering, se decidió mantener como categorías individuales solo las marcas más frecuentes y agrupar el resto en otros. Esto reduce el ruido de las marcas raras sin perder información significativa sobre las marcas dominantes

```python
print(df["marca"].nunique())
print(df["marca"].value_counts())
```

```python
print(df["marca"].value_counts()/len(df["marca"])*100)
```

```python
total_registros = len(df)
umbral = 0.01 * total_registros  # 1% del total
print(f"Total registros: {total_registros}, Umbral (1%): {umbral}")

marcas_frecuentes = df['marca'].value_counts()
marcas_a_mantener = marcas_frecuentes[marcas_frecuentes >= umbral].index.tolist()
print(f"Marcas a mantener: {marcas_a_mantener}")
print(f"Cantidad de marcas a mantener: {len(marcas_a_mantener)}")

df['marca_agrupada'] = df['marca'].apply(lambda x: x if x in marcas_a_mantener else 'OTROS')
df.drop(columns=['marca'], inplace=True)
print("\nDistribución de 'marca_agrupada':")
print(df['marca_agrupada'].value_counts(normalize=True) * 100)
```

### Prueba de winsorizacion

```python
df_winsorizado = df.copy()

p95_usd = df_winsorizado['valor_referencial_2025_pen'].quantile(0.95)
df_winsorizado['valor_referencial_2025_pen'] = df_winsorizado['valor_referencial_2025_pen'].clip(upper=p95_usd)

p95_soles = df_winsorizado['valor_adquisicion_pen'].quantile(0.95)
df_winsorizado['valor_adquisicion_pen'] = df_winsorizado['valor_adquisicion_pen'].clip(upper=p95_soles)

print("Winsorización completada - valores limitados al percentil 95")
print(f"Valor máximo de referencia después de winsorización: {df_winsorizado['valor_referencial_2025_pen'].max()}")
print(f"Valor máximo de adquisicion después de winsorización: {df_winsorizado['valor_adquisicion_pen'].max()}")
```

```python
# Comparación de estadísticas descriptivas
print("=== ESTADÍSTICAS COMPARATIVAS ===")
print("\nDatos originales limpios:")
print(df[['valor_referencial_2025_pen', 'valor_adquisicion_pen']].describe())

print("\nDatos winsorizados:")
print(df_winsorizado[['valor_referencial_2025_pen', 'valor_adquisicion_pen']].describe())
```

Aqui se aplica Winsorización solo para ver los efectos, no se tomara en cuenta para el desarrollo puesto que los valores son extremos, tienen sentido dada la lógica del negocio (vehiculos de lujo o raros con precios muy altos) y no se pueden reemplazar
